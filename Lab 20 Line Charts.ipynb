### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
df['Age_num'] = df['Age'].str.extract(r'(\d+)', expand=False).astype(float)

bins = [0, 18, 24, 34, 44, 54, 64, 100]
labels = ['<18','18–24','25–34','35–44','45–54','55–64','65+']
df['AgeGroup'] = pd.cut(df['Age_num'], bins=bins, labels=labels, right=False)

median_comp = df.groupby('AgeGroup', observed=False)['ConvertedCompYearly'].median().reset_index()

plt.figure(figsize=(10,6))
plt.plot(median_comp['AgeGroup'], median_comp['ConvertedCompYearly'], marker='o', linestyle='-', color='blue')

plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation (Yearly)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
def extract_age(age_str):
    try:
        return float(age_str.split()[0])
    except:
        return None

df['Age_num'] = df['Age'].apply(extract_age)

df_age = df[(df['Age_num'] >= 25) & (df['Age_num'] <= 45)].copy()

bins = [25, 29, 34, 39, 45]
labels = ['25–29','30–34','35–39','40–45']

df_age['AgeGroup'] = pd.cut(df_age['Age_num'], bins=bins, labels=labels, right=True)

median_comp = df_age.groupby('AgeGroup', observed=True)['ConvertedCompYearly'].median().reindex(labels)

median_comp = median_comp.fillna(0)

plt.figure(figsize=(8,5))
plt.plot(labels, median_comp, marker='o', linestyle='-', color='green')
plt.title('Median Yearly Compensation for Ages 25–45')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation (Yearly)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
df['YearsCodePro'].head()

def parse_experience(exp):
    try:
        if exp.startswith('Less than'):
            return 0.5
        elif exp.startswith('More than'):
            return 51  # or some upper limit
        else:
            return float(exp)
    except:
        return None

df['Experience'] = df['YearsCodePro'].apply(parse_experience)

bins = [0, 2, 5, 10, 20, 30, 50, 60]  
labels = ['<2','2–5','5–10','10–20','20–30','30–50','50+']

df['ExperienceGroup'] = pd.cut(df['Experience'], bins=bins, labels=labels, right=True)

median_job_sat = df.groupby('ExperienceGroup', observed=True)['JobSatPoints_6'].median().reindex(labels)

plt.figure(figsize=(8,5))
plt.plot(labels, median_job_sat, marker='o', linestyle='-', color='blue')
plt.title('Median Job Satisfaction (JobSatPoints_6) by Experience Level')
plt.xlabel('Experience Level (Years)')
plt.ylabel('Median Job Satisfaction Score')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
def parse_experience(exp):
    try:
        if pd.isna(exp):
            return None
        if exp.startswith('Less than'):
            return 0.5
        elif exp.startswith('More than'):
            return 51
        else:
            return float(exp)
    except:
        return None

df['Experience'] = df['YearsCodePro'].apply(parse_experience)

bins = [0, 2, 5, 10, 20, 30, 50, 60]  
labels = ['<2','2–5','5–10','10–20','20–30','30–50','50+']

df['ExperienceGroup'] = pd.cut(df['Experience'], bins=bins, labels=labels, right=True)

median_comp = df.groupby('ExperienceGroup', observed=True)['ConvertedCompYearly'].median().reindex(labels)

plt.figure(figsize=(8,5))
plt.plot(labels, median_comp, marker='o', linestyle='-', color='purple')
plt.title('Median ConvertedCompYearly by Experience Level')
plt.xlabel('Experience Level (Years)')
plt.ylabel('Median Yearly Compensation (ConvertedCompYearly)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
