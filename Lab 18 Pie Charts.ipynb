### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
databases = df['DatabaseWantToWorkWith'].dropna().astype(str).str.split(';').explode()

top5_databases = databases.value_counts().head(5)
print("Top 5 Databases:\n", top5_databases)

plt.figure(figsize=(6,6))
plt.pie(top5_databases, labels=top5_databases.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top 5 Databases Developers Want to Work With')
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
dev_types = df['DevType'].dropna().astype(str).str.split(';').explode()

top5_dev_types = dev_types.value_counts().head(5)
print("Top 5 Developer Types:\n", top5_dev_types)

plt.figure(figsize=(6,6))
plt.pie(top5_dev_types, labels=top5_dev_types.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top 5 Developer Roles')
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
op_systems = df['OpSysProfessional use'].dropna().astype(str).str.split(';').explode()

top5_os = op_systems.value_counts().head(5)
print("Top 5 Operating Systems:\n", top5_os)

plt.figure(figsize=(6,6))
plt.pie(top5_os, labels=top5_os.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top Operating Systems Used Professionally')
plt.show()
```

### Task 2: Additional Visualizations and Comparisons
